```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

#!pip install sweetviz
import sweetviz
```

```python
df=pd.read_csv('Loan_Data.csv').dropna()
```

```python
df
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
df['Education']=le.fit_transform(df['Education'])
df['Married']=le.fit_transform(df['Married'])
df['Gender']=le.fit_transform(df['Gender'])
df['Loan_Status']=le.fit_transform(df['Loan_Status'])
df['Self_Employed']=le.fit_transform(df['Self_Employed'])
df['Property_Area']=le.fit_transform(df['Property_Area'])
df['Dependents']=le.fit_transform(df['Dependents'])
```

```python
#train test split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
fs=StandardScaler()
X_train=fs.fit_transform(X_train)
X_test=fs.transform(X_test)
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel='linear',C=1,degree=3,random_state=0)
classifier.fit(X_train,Y_train)
classifier.score(X_test,Y_test)
```

```python
classifier=SVC(kernel='rbf',C=1,degree=3,random_state=0)
classifier.fit(X_train,Y_train)
classifier.score(X_test,Y_test)
```

```python
classifier=SVC(kernel='sigmoid',C=1,degree=4,random_state=0)
classifier.fit(X_train,Y_train)
classifier.score(X_test,Y_test)
```
